```python

    
def meanvector(vector):
    row_mean_vector = np.mean(vector, axis = 1)
    #print(row_mean_vector)
    row_sum = vector.sum(axis = 0)
    #print(row_sum)
    mean_vector = row_sum*(1/len(row_mean_vector))
    return mean_vector
    

###function that truncates image data into the first n rows





def slice(matrix,n):
    sliced_matrix = matrix[:,range(n)]
    return sliced_matrix



```
